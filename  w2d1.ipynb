#### Performing Calculations on ints and floats

##### Addition

```python
num1 = 2
num2 = 5.4

result = num1 + num2

print(result)

result += 2 #this will redefine the "result" variable
print(result)
```

##### Subtraction

```python
result_diff = num2 - num1
print(result_diff)

result_diff -= 1
print(result_diff)

```

##### Multiplication

```python
result_prod = num1 * num2
print(result_prod)

result_prod*= 2
print(result_prod)
```

##### Division

```python
result_div = num2 / num1
print(result_div)

result_div /=2
print(result_div)
```

##### Floor Division

```python
result_floor = num2 //num1 #opposite of modulo, drops all values after the decimal
print(result_floor)

result_floor //= 2
print(result_floor)
```

##### Modulo

```python
result_mod = 33 % 5 #gives you remainder after division
print(result_mod)

result_mod %= 2
print(result_mod)
```

### Elif Statements

```python
some_num = 4

if some_num % 2 == 1:
    print('even number')
elif some_num > 0:      #elif only runs if the "if" statement above it runs as false
    print("poditive number")
```

### Else Statements

```python
some_num = 0

if some_num % 2 == 1:
    print('even number')
elif some_num > 0:      #elif only runs if the "if" statement above it runs as false
    print("poditive number")
else:
    print("num is 0")
```

### For Loops

```python
# for keyword, variable counter, in keyword, iterable

name = "Brian Stanton"

for letter in name:
    print("hello")
    print(letter)
    
print(name[4])
    
    
for i in range(len(name)):
    print(i)
```

##### Using 'in' keyword

```python
# see above

for letter in name:
    if letter =="i":
        print(letter)
        
for index in range(len(name)):
    if name[index] =="i":
        print(index)
```

##### Continue Statement

##### Break Statement

```python
# will break out of current loop
```

```python
for i in range(10):
    if i == 5:
        break
    print(i)
```

##### Pass Statement

```python
# mostly used as a placeholder, and will continue on same iteration
```

```python
for i in range(10):
    if i == 5:
        pass
    print(i)
```
